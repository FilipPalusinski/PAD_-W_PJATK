```python
#ZADANIE 1
import pandas as pd 
import numpy as np
data = pd.read_csv('president_heights.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
mean = np.mean(heights)
print(mean)
```

```python
std = np.std(heights)
print(std)
```

```python
min = np.min(heights)
max = np.max(heights)
print(f"min = {min}, max = {max}")
```

```python
print("Mean height:       ", mean) 
print("Standard deviation:", std)
print("Minimum height:    ", min) 
print("Maximum height:    ", max)
```

```python
print("25th percentile:   ", np.percentile(heights, 25))
print("Median:            ", np.median(heights))
print("75th percentile:   ", np.percentile(heights, 75))
```

```python
#ZADANIE 3
import numpy as np
import pandas as pd
# use pandas to extract rainfall inches as a NumPy array
rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values 
inches = rainfall / 254.0  # 1/10mm -> inches 
inches.shape
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import seaborn; seaborn.set()  
# set plot styles 
plt.hist(inches, 40)
```

```python
daysWithoutRain = np.sum(inches == 0)
daysWithRain = np.sum(inches > 0)
x = np.sum(inches > 0.5)
y = np.sum((inches < 0.2) & (inches > 0))
```

```python
print("Number days without rain:      ", daysWithoutRain)
print("Number days with rain:         ", daysWithRain)
print("Days with more than 0.5 inches:", x)
print("Rainy days with < 0.2 inches  :", y)
```

```python
medianPrecipInRainDays = np.median(inches[inches > 0])

summerStartAt = 172
summerEndAt = 262

summerDays = inches[summerStartAt-1:summerEndAt] #jeden element mniej, bo tablica zaczyna sie od elementu 0
medianPrecipInSummerDays = np.median(summerDays)
maxPrecipInSummerDays = np.max(summerDays)

nonSummerDays = np.concatenate([inches[:summerStartAt-1], inches[summerEndAt:]])
maxPrecipInNonSummerDays = np.max(nonSummerDays)
```

```python
print("Median precip on rainy days in 2014 (inches): ", medianPrecipInRainDays) 
print("Median precip on summer days in 2014 (inches):  ", medianPrecipInSummerDays) 
print("Maximum precip on summer days in 2014 (inches): ", maxPrecipInSummerDays) 
print("Maximum precip on non-summer rainy days (inches):", maxPrecipInNonSummerDays) 
#tu był błąd, "median" zamieniłem na "maximum", ponieważ tak było w poleceniu
```
